```python
#Find a way to limit reviews
#Find documentation 
us_reviews = reviews_all(
    'com.joytunes.simplypiano',
    sleep_milliseconds=0, # defaults to 0
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.MOST_RELEVANT
)
```

https://towardsdatascience.com/generate-meaningful-word-clouds-in-python-5b85f5668eeb

```python
review_content = df_jt.content.tolist()
review_content = ''.join(review_content).lower()
```

```python
#pip install wordcloud in environment
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
stop_words= ['piano','play','app','keyboard','give','got','u','playing','app,','piano.','im','app.','download'] + list(STOPWORDS)
wordcloud = WordCloud (stopwords= stop_words, background_color = 'white', max_words = 300).generate(review_content)
```

```python
plt.figure(figsize=(15,10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
all_words = review_content.lower().split()
clean_words = [w for w in all_words if not w in stop_words]
freq_clean = nltk.FreqDist(clean_words)
clean_dict = freq_clean.items()
print(clean_dict)
```

https://stackoverflow.com/questions/613183/how-do-i-sort-a-dictionary-by-value

Dictionaries can't be sorted because they are ordered, but you can make a sorted list?

sorted(d.items(), key=lambda x: x[1])
This will sort the dictionary by the values of each entry within the dictionary from smallest to largest.
Reverse=True for descending

```python
sorted(freq_clean.items(), key = lambda x:x[1], reverse = True)
```
